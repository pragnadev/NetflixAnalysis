```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["release_year"] , bins = 30 , kde = True )
plt.title("distribution of realease years")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="type" , data = df , palette= "coolwarm")
plt.title("count of movies vs tv shows")
plt.show()
```

```python
from collections import Counter
genres = df["listed_in"].str.split(",").explode()
top_genres = genres.value_counts().head(10)
plt.figure(figsize=(10,15))
sns.barplot(x=top_genres.values , y= top_genres.index , palette="coolwarm")
plt.title("Top 10 most common netflix generes")
plt.xlabel("count")
plt.ylabel("genre")
plt.show()
```

```python
df_movies = df[df["type"] == "Movie"].copy()
df_movies["duration"] = df_movies["duration"].str.replace(" min", "").astype(float)  # Convert to numeric

plt.figure(figsize=(8,4))
sns.histplot(df_movies["duration"], bins=30, kde=True, color="red")
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```
